```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis

import bs4
from bs4 import BeautifulSoup
import lxml
import html5lib
import requests
```

```python
# Get the webpage content into HTML format
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
# Parse with Beautiful Soup
soup = BeautifulSoup(source,'lxml')
# get all tables
tables = soup.find_all("table")
# first table
table = tables[0]
df = pd.DataFrame()
rows  = table.find_all("tr")
# extract column name from first row
column_names = []
header_cells = rows[0].find_all("th")
header_cells

for cell in header_cells:
    header = cell.text
    header = header.strip()
    column_names.append(header)
column_names
```

```python
# get data from table
rowdata = []
for row in rows[2:]:
    #create an empty tuple
    dt  = ()
    cells = row.find_all("td")
    for cell in cells:
        font = cell.find("font")
        if font is not None:
            celldata = font.text
        else:
            celldata = cell.text
        celldata = celldata.strip()
        celldata = celldata.replace("\n"," ")
        dt = dt+(celldata,)
    rowdata.append(dt)
df = pd.DataFrame(rowdata, columns = column_names)
df
```

```python
df1 = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).reset_index()
df1
```

Import map rendering library and map co-ordinates

```python
!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(new_df['Latitude'], new_df['Longitude'], new_df['Borough'], new_df['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
